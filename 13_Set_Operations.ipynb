# Set Operations

```python
import numpy as np
```

## Set Operations

```python
arr = np.array([1,2,3,4])
arr2 = np.array([2,4,5,6])

print("Array 1:", arr)
print("Array 2:", arr2)

# Union - combine all unique elements
print("\nUnion (all unique):", np.union1d(arr, arr2))

# Intersection - elements in both arrays
print("Intersection (common):", np.intersect1d(arr, arr2))

# Difference - elements in arr1 but not in arr2
print("Difference (arr1 - arr2):", np.setdiff1d(arr, arr2))

# Symmetric difference - elements in either but not both
print("Symmetric difference:", np.setxor1d(arr, arr2))
```

## Performance with assume_unique Parameter

```python
import time

arr1 = np.array([1,2,3,4])
arr2 = np.array([2,5,6,4])

# Without assume_unique (slower)
start = time.time()
result = np.intersect1d(arr1, arr2)
end = time.time()
print(f"Without assume_unique: {result}, Time: {end-start:.6f}s")

# With assume_unique (faster)
start = time.time()
result = np.intersect1d(arr1, arr2, assume_unique=True)
end = time.time()
print(f"With assume_unique: {result}, Time: {end-start:.6f}s")
```
