# Copy vs View

Copy:
The copy owns the data and any changes made to the copy will not affect original array, and any changes made to the original array will not affect the copy.

View:
The view does not own the data and any changes made to the view will affect the original array, and any changes made to the original array will affect the view.



Copy

to copy the data from another array we will use array_created_by_us.copy()    function

```python

arr = np.array([1,2,3,4,5])
x = arr.copy()
#after copying of arr to x we can add an element to arr but the element won't come in x
arr[0] = 9
print(arr)
print(x)
print('\n')
```

```python
# checking whether x is a view or a copy
print(x.base)
 # if the array was created by using using copy it gives none in output
```

View 

To view the data from another array we will use array_created_by_us.view()  function

Using a View saves memory because it doesn't create a new array

```python
#when we use  .view the element which was added to arr also comes to x
x = arr.view()
arr[0] = 9
print(arr)
print(x)
```

```python
# checking whether x is a view or a copy
print(x.base)
 # if the array was created by using using view it gives original array in output
```

# NaN Handling

NaN stands for not a number
it is a float values for missing values

```python
# here we are a an array with NaN values and we will see how to handle them
arr = np.array([1, 2, np.nan, 4])
print(arr)
```

```python
# checking for Nan values in an array if there is a NaN value in an array it gives True in the output else False
# for that we will use np.isnan() function
np.isnan(arr)
```

```python
# Filtering out NaN values from the array
# for this we will use ~np.isnan() function
# so it gives non Nan values 
filtered_arr = arr[~np.isnan(arr)]
print(filtered_arr)
```

```python
# counting number of NaN values
# here we are sum function also because it add all the NaN values in an array
num_nan = np.isnan(arr).sum()
print(num_nan)
```

```python
# Example
arr = np.array([1, 2, np.nan, 4, np.nan,6])
num_nan = np.isnan(arr).sum()
print(num_nan)
```

```python
# replacing Nan values with some other values 
# np.nan_to_num(array want to replace NaN values, nan=value we want to replace) 
arr = np.array([1, 2, np.nan, 4, np.nan,6])
x = nan_to_num = np.nan_to_num(arr, nan=0)
print(x)
# so if we see the output all the Nan Values are replaced with 0
```

Aggregation(forming a number) that ignore naN values



```python
# here these values ignore naN values and continues with other values
print(np.nanmean(arr))
print(np.nansum(arr))
print(np.nanmax(arr))
print(np.nanmedian(arr))
```
