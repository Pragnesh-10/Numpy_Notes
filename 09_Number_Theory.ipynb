# Number Theory Functions

```python
import numpy as np
```

## LCM (Least Common Multiple)

```python
num1 = 4
num2 = 6

print(f"LCM of {num1} and {num2}:", np.lcm(num1, num2))
print(f"GCD of {num1} and {num2}:", np.gcd(num1, num2))
```

## LCM and GCD of Array Elements

```python
# Using reduce() to apply LCM/GCD across array
arr = np.array([3, 4, 5])
print("Array:", arr)
print("LCM of all elements:", np.lcm.reduce(arr))
print("GCD of all elements:", np.gcd.reduce(arr))
```

```python
# Range of numbers
arr = np.arange(1, 11)
print("Array (1-10):", arr)
print("LCM of 1-10:", np.lcm.reduce(arr))
print("GCD of 1-10:", np.gcd.reduce(arr))
```
