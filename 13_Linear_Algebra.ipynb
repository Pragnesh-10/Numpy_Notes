## Creating Matrices

```python
# Creating a matrix from string
matrix_str = np.matrix('1 2; 3 4')
print("Matrix from string:")
print(matrix_str)
```

```python
matrix_str = np.matrix('1 2; 3 4; 5 6')
print(matrix_str)
```

```python
# Creating a matrix from array
array_data = np.array([[1, 2], [3, 4]])
# here first we have created an array and converted into numpy matrix by using function np.matrix
matrix_from_array = np.matrix(array_data)
print("\nMatrix from array:")
print(matrix_from_array)
```

```python
print(type(array_data))
```

```python
print(type(matrix_from_array)) # datatype of matrix
```

```python
# Using asmatrix()
array_data = np.array([[5, 6], [7, 8]])
matrix_data = np.asmatrix(array_data)
print("Matrix via asmatrix:")
print(matrix_data)
print("Type array:", type(array_data))
print("Type matrix:", type(matrix_data))
```

## Matrix Addition 

```python
# Addition of matrices. we can add matrices in two different ways by using + operator and add() function
# Adding two matrices using the + operator
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = A + B

print(C)
```

```python
# Adding two matrices using add() function
c = np.add(A,B)
print(c)
```

# Matrix Subtraction

```python
# subtraction of two matrices. we can subtract matrices in two different ways by using - operator and subtract() function
# Subtracting two matrices using the - operator
A = np.array([[5, 8], [7, 10]])
B = np.array([[2, 4], [3, 6]])
C = A - B
print(C)
```

```python
# subtracting matrix using subtract() function
c = np.subtract(A,B)
print(c)
```

## Matrix Multiplication

### Dot Product

```python
# Dot product of the two arrays
# by using function dot()
a = np.array([[1,2],[3,4]])
b = np.array([[11,12],[13,14]])
print(np.dot(a,b))
# [[1*11+2*13, 1*12+2*14],[3*11+4*13, 3*12+4*14]]
```

```python
# by using function @ operator
dot = a @ b
print(dot)
```

### Vector Product

```python
a = np.array([[1,2],[3,4]])
b = np.array([[11,12],[13,14]])

# Vector dot product (1*11 + 2*12 + 3*13 + 4*14)
result = np.vdot(a, b)

print(result)
```

### Inner Product

```python
a = np.array([[1,2],[3,4]])
b = np.array([[11,12],[13,14]])

result = np.inner(a, b)
print("Inner product:")
print(result)
# [[ 1*11+2*12, 1*13+2*14]
#   [3*11+4*12, 3*13+4*14]]
```

```python
# without using matlib function we can directly mix  2-D mixed with 1-D
a = np.array([[1,0],[0,1]])
b = np.array([1,2])

print(a @ b)
```

```python
print(b @ a)
```

## Matrix Determinant

```python
# the determinant of the array

a = np.array([[1,2], [3,4]])
print(np.linalg.det(a))
# 1*4-2*3
```

```python
# the determinant of the array or matrix 
# the result will be same for both array and matrix

b = np.array([[6,1,1], [4, -2, 5], [2,8,7]])
print(b)
print(np.linalg.det(b))
# 6*(-2*7 - 5*8) - 1*(4*7 - 5*2) + 1*(4*8 - -2*2)
print(6*(-2*7 - 5*8) - 1*(4*7 - 5*2) + 1*(4*8 - -2*2))
```

```python
# here i have converted above array into matrix then i found det of a matrix 
# therefore the result will be same for both array and matrix
x = np.asmatrix(b)
print(x)
print(np.linalg.det(x))
```

## Matrix Inverse

```python
A = np.array([[1, 2], [3, 4]])
print("Original Matrix:")
print(A)

A_inv = np.linalg.inv(A)
print("\nInverse Matrix:")
print(A_inv)
```

```python
# Verify inverse (A × A_inv = Identity)
identity = np.dot(A, A_inv)
print("A × A_inv:")
print(identity)

# Check if it's identity matrix
is_identity = np.allclose(identity, np.eye(2))
print("\nIs identity matrix?", is_identity)
```
