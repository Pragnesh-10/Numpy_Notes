# Trigonometric Functions

```python
import numpy as np
```

## Basic Trigonometric Functions

```python
# sin, cos, tan
result = np.sin(np.pi/2)
print("sin(π/2):", result)

# Hyperbolic functions
print("sinh(π/2):", np.sinh(np.pi/2))
print("cosh, tanh are also available")
```

## Inverse Trigonometric Functions

```python
# arcsin returns radians
x = np.arcsin(1)
print("arcsin(1) in radians:", x)

# Convert to degrees
y = np.degrees(x)
print("Converted to degrees:", y)

# Convert to radians
z = np.radians(90)
print("90 degrees in radians:", z)
```

## Degree and Radian Conversion

```python
# Degrees to radians
arr = np.array([90, 180, 270, 360])
print("Degrees:", arr)
x = np.deg2rad(arr)
print("In radians:", x)

# Radians to degrees
arr = np.array([np.pi/2, np.pi/3, np.pi/4])
print("\nRadians:", arr)
print("In degrees:", np.rad2deg(arr))
```

## Finding Angles

```python
# arcsin for single value
x = np.arcsin(1)
print("arcsin(1):", x)
print("arcsinh(1):", np.arcsinh(1))
```

```python
# arcsin for array
arr = np.array([-1, 0, 1])
print("Array:", arr)
print("arcsin:", np.arcsin(arr))
print("arccos and arctan also available")
```

## Pythagorean Theorem

```python
# Calculate hypotenuse
base = 3
height = 4
hyp = np.hypot(base, height)
print(f"Hypotenuse of {base}-{height} triangle:", hyp)
```
